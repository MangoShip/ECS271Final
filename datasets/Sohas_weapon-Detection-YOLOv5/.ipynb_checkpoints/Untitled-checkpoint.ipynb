```python
import cv2
import os, sys
from skimage import io
```

```python
def getBboxFromDarknet(darknetLine, frameWidth, frameHeight):
    xCenter = float(darknetLine.split(' ')[1])
    yCenter = float(darknetLine.split(' ')[2])
    w = float(darknetLine.split(' ')[3])
    h = float(darknetLine.split(' ')[4])

    x1 = int((xCenter - w/2) * frameWidth)
    y1 = int((yCenter - h/2) * frameHeight)
    x2 = int((xCenter + w/2) * frameWidth)
    y2 = int((yCenter + h/2) * frameHeight)

    return x1,y1,x2,y2
```

```python
imgPath = 'obj_train_data/images/test/KravMagaKnifeDefenseTechniques023.jpg'
img = cv2.imread(imgPath)
h, w = img.shape[:2]
with open(imgPath.replace('images', 'labels').replace('jpg', 'txt'), 'r') as labelTxt:
    darkLine = labelTxt.read()
x1, x2, y1, y2 = getBboxFromDarknet(darkLine, w, h)
img = cv2.rectangle(img, (x1, y1), (x2, y2), (0,255,0), 2)
```
